## [1. Professional and Legitimate Python Code](#starter)


Python is famous for being a "readable" language, but that readability is enforced by some rules.


* **Indentation** (Required "Golden Rule")

    Indentation in Python is mandatory and defines the scope of code blocks. Unlike many other programming languages where indentation is used primarily for readability, Python relies on indentation to determine program structure.

    - *Definition:* Indentation refers to the spaces or tabs at the beginning of a line of code.

    - *Consistency:* Tabs and spaces must not be mixed. In Python 3, doing so will result in an error.

    - *Block Structure:* Lines with the same level of indentation belong to the same block of code.

    - *Common Error:* Improper indentation can lead to errors such as  
      `IndentationError: unexpected indent`



* **The "Colon" Rule** (Required)

    In Python, a colon (`:`) must appear at the end of any statement that introduces a new indented block. The colon signals that the following lines will form a block of code.

    - *Function Definitions:*   `def function_name():`

    - *Conditional Statements:*   `if condition:`

    - *Loops:*   `for item in list:`


* **Spacing** (Suggested)

    Proper spacing improves code readability and maintains consistency with Python style guidelines.

    - *Horizontal Spacing:* Use one space around assignment (`=`) and comparison operators (`==`, `>`).

    - *Vertical Spacing:* Use two blank lines between function and class definitions. Use one blank line to separate logical sections within a function.




* **Documentation** (Suggested)

  Use *docstrings* to describe the purpose and behavior of functions, and *comments* to explain *why* the code exists rather than *what* it does.

    * A *docstring* (documentation string) is a string literal used to document a module, class, function, or method. It is written immediately after the definition and is enclosed in triple quotes (`"""`). Docstrings can be accessed programmatically and are commonly used by documentation tools.

    * A *comment* is a line of text preceded by the hash symbol (`#`). Comments are ignored by the Python interpreter and are used to clarify intent, assumptions, or reasoning behind a piece of code.

  ```python
  def calculate_growth(initial_value, rate):
      """
      Calculates the exponential growth over a single period.
      """
      # Apply standard growth formula
      new_total = initial_value * (1 + rate)

  ```

````
# Version 2: Does this code cell execute correctly? Why or why not?

'doesn\'t'
````

- Why `'doesn't'` fails but `'doesn\'t'` works

  - `'doesn't'` fails:
    - Single quotes `'` delimit the string.
    - The `'` in `doesn't` is interpreted as the end of the string.
    - The remaining `t'` is invalid  **syntax error**.

  - `'doesn\'t'` works:
    - Backslash `\` escapes the apostrophe.
    - Interpreter treats `\'` as a literal `'` inside the string.
    - String ends correctly at the final `'`.

  - Alternative: double quotes
    - `"doesn't"` works without escaping.
    - Single quote inside double-quoted string is allowed.

## [4. Working with Variables and Expressions](#variables)

* Python is an object oriented programming language.

  Almost everything in Python is an object, with its properties and methods.

  * *Object*: Everything in Python is an object (numbers, strings, lists, functions, classes).

    * Has a *type* (e.g., `int`, `list`, `str`)
    * Has a *value*
    * Lives somewhere in *memory*
    * Objects exist independently of variable names.

  * *Variable (Name)*

    * A variable is *just a name*, not a container.
    * Variables do *not store values* directly but *bind* to (points to) an object.
    * Assignment (`=`) creates or changes this binding.

  * *Reference*

    * A reference is the link between a variable and an object.
    * Variables reference objects, not copies of them.
    * Multiple variables can reference the *same object*.

### 4.0  Assignment

* *Definition and Purpose*
  * *Reference by Name*: Variables allow programmers to use descriptive names (like `age` or `birth`) to refer to data, rather than abstract memory locations.
  * *Storage*: They act as containers for values that can be retrieved or changed throughout a program.

* *The Assignment Operator (=)*
  * *Syntax*: Uses the `=` symbol to assign a value to a variable.
  * *Logic*: The variable name must be on the **left**, and the value being assigned must be on the **right**.
  * *Distinction from Mat**: In programming, `=` does not mean "is equal to" in a mathematical sense; it is an action that *sets* the value.

* *Name Binding*

    * Assignment binds a name to an object.
    * Example:
      * `x = 10` binds `x` to the object `10`, or assigns the integer 10 to the variable named **x**
      * `x = 20` rebinds `x` to a new object `20`


### 4.1 Naming Rules



* **Naming Essentials**
  * *Permitted Characters*: Use only letters, digits, and underscores.
  * *No Leading Digits*: A name must begin with a letter or an underscore, never a number (e.g., ``2nd_place`` is illegal).
  * *Case Sensitivity*: Case matters; `Total` is distinct from `total`.
  * *Reserved Words*: You cannot use Python's "reserved words" (like class, def, or True) because they are already assigned to specific programming functions.


      | | | | | |
      | :--- | :--- | :--- | :--- | :--- |
      | False | await | else | import | pass |
      | None | break | except | in | raise |
      | True | class | finally | is | return |
      | and | continue | for | lambda | try |
      | as | def | from | nonlocal | while |
      | assert | del | global | not | with |
      | asynch | elif | if | or | yield |

* **Stylistic Recommendations**
  * *Snake Case:* It is recommended to use all lowercase letters with underscores between words (e.g., current_temperature).

  * *Clarity Over Brevity:* Use words instead of single characters (e.g., `count` instead of `c`).

  * *Purpose-Driven:* Choose names that clearly indicate what the variable represents to make the code easier for others (and yourself) to read.

  * Some naming conventions
  | Item | Style | Example |
  | :--- | :--- | :--- |
  | **Variables/Functions** | `snake_case` | `total_score`, `get_data()` |
  | **Constants** | `ALL_CAPS` | `MAX_THRESHOLD = 100` |
  | **Classes** | `PascalCase` | `UserAccount` |

* Variables in Python are not buckets containing things; they're pointers (they point to objects).

- Objects have:
  - *Identity*  where they live in memory
  - *Type*  what kind of object it is
  - *Value*  the data it holds

    ```python
    x = [1, 2, 3]
    id(x)    # returns the memory address of the list object
    type(x)  # <class 'list'>
    ```

- Variables are references (pointers)

   - Variables do not store the object itself.

   - They store a reference (pointer) to the object.


- Mutability matters

   - Mutable objects (lists, dicts, sets): **Can** be changed in-place. Changes via any reference affect the same object.

- Immutable objects (tuples, strings, numbers): **Cannot** be changed in-place.

## [6. Built-in Functions vs Methods in Python](#methods)

- Operators

  - Definition: Special symbols that perform operations on values.

  - Can work on **multiple types**, and often call special methods internally.

  - Examples: ``+``, ``-``, ``*``, ``/``, ``==``, ``>``, ``and``, ``or``


- Built-in Functions
  - General functions provided by Python that work on **multiple data types** or independently.
  - Called **directly**, not tied to a variable.
  - Examples:
    ```python
    len("hello")    # 5
    max([1, 3, 2])  # 3
    sum([1, 2, 3])  # 6
    type(123)       # <class 'int'>


- Methods

  - Functions that belong to a **specific object or class**.

  - Called using dot notation (object.method()).

  - Can access or modify the object it belongs to.

  - Examples: list.append(), str.upper(), dict.get()
