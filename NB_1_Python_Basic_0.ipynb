## [1. "Essence" of Coding](#whatCoding)
Coding is the process of providing a computer with a set of precise, unambiguous instructions to solve a problem or perform a task.
Computers are incredibly fast, obedient but ``unthinking" in that they can only do exactly what they're told, which makes logical thinking and meticulous instructions paramount in coding.

### 1.1. The Computer as an Executor
The fundamental architecture of almost any computational task is the **Input** -> **Process** -> **Output** cycle.
* **Input:** Data or information the computer receives, such as text from a file, numbers typed by a user, or sensor readings.
* **Processing:** The computer performs operations on the input according to your instructions, like calculations, comparisons, or data manipulation.
* **Output:** The result of the processing, presented back to the user or another system, such as a number displayed on a screen or a saved file.



### 1.2. Algorithms: The Recipe for Code
An **algorithm** is a step-by-step procedure or a set of rules used to solve a specific problem. Before you write any code, you typically design an algorithm. Your algorithms are then translated into a language the computer can understand, such as Python or R.

Example Algorithm: Making a Cup of Tea ([Bouras, A.S. The Algorithm for Making a Cup of Tea.](https://www.bouraspage.com/repository/algorithmic-thinking/the-algorithm-for-making-a-cup-of-tea))

1.  Put the teabag in a cup.
2.  Fill the kettle with water.
3.  Boil the water in the kettle.
4.  Pour some of the boiled water into the cup.
5.  Add milk to the cup.
6.  Add sugar to the cup.
7.  Stir the tea.
8.  Drink the tea.



In coding, your algorithms are translated into a language the computer can understand (Python, R, MATLAB, Java, C++, etc.).




### 1.3. Programming Languages: the grammar of code

1. Syntax  Rules of form (how code is written).
2. Semantics  Rules of meaning (what code does).
3. Pragmatics  Rules of use (how code is best written in real contexts).
4. Lexicon & Grammar  The building blocks that make up syntax.
5. Type System & Execution Rules  Constraints that define correctness at runtime.

### 2.3. Variables and Data Types
**Variables** are names given to memory locations that store data. Think of them as labeled boxes where you can put different kinds of information.
**Data Types** define the kind of data a variable can hold (e.g., whole numbers, decimal numbers, text, true/false values).

Python is dynamically typed, meaning you don't declare the type explicitly. R: Has types like `numeric`, `integer`, `character`, `logical`, `list`, `data.frame`.


````python
# Integer (whole number)
age = 30
print(f"Age: {age}, Type: {type(age)}")

# Float (decimal number)
price = 19.99
print(f"Price: {price}, Type: {type(price)}")

# String (text)
name = "Alice"
print(f"Name: {name}, Type: {type(name)}")

# Boolean (True/False)
is_active = True
print(f"Is Active: {is_active}, Type: {type(is_active)}")

# List (ordered collection, mutable)
numbers = [1, 2, 3, 4, 5]
print(f"Numbers: {numbers}, Type: {type(numbers)}")

# Dictionary (key-value pairs, unordered, mutable)
person = {"name": "Bob", "age": 25}
print(f"Person: {person}, Type: {type(person)}")
````

### 2.4. Operators
Operators are special symbols or keywords that perform operations on values or operands. They are fundamental to performing calculations, comparisons, and logical evaluations in any programming language.

Common types of operators include:
* **Arithmetic:** For calculations (`+`, `-`, `*`, `/`).
* **Comparison (Relational):** For comparing values (`==`, `!=`, `<`, `>`).
* **Logical:** For combining boolean conditions (`and`, `or`, `not`).
* **Assignment:** For assigning values to variables (R: `<-`; Python `=`).

Note: In R, both `=` and `<-` assign values, but `<-` is preferred in most R code because it avoids confusion with the `=` when `=` is used for function arguments. For example `seq(from = 1, to = 10, by = 2)`

**Python:**  Operators in python:



````python
# Arithmetic Operators
x, y = 15, 4
print(f"Arithmetic:")
print(f"  Addition (x + y): {x + y}")
print(f"  Subtraction (x - y): {x - y}")
print(f"  Multiplication (x * y): {x * y}")
print(f"  Division (x / y): {x / y}")
print(f"  Floor Division (x // y): {x // y}")
print(f"  Modulo (x % y): {x % y}")
print(f"  Exponentiation (x ** y): {x ** y}")

# Comparison Operators
a, b = 10, 20
print(f"\nComparison:")
print(f"  Equal (a == b): {a == b}")
print(f"  Not Equal (a != b): {a != b}")
print(f"  Less than (a < b): {a < b}")
print(f"  Greater than (a > b): {a > b}")
print(f"  Less than or Equal (a <= b): {a <= b}")
print(f"  Greater than or Equal (a >= b): {a >= b}")

# Logical Operators
is_sunny = True
is_warm = False
print(f"\nLogical:")
print(f"  AND (is_sunny and is_warm): {is_sunny and is_warm}")
print(f"  OR (is_sunny or is_warm): {is_sunny or is_warm}")
print(f"  NOT (not is_sunny): {not is_sunny}")

# Assignment Operators
num = 5
print(num)
print(f"\nAssignment:")
print(f"  Initial num: {num}")
num += 3 # num = num + 3
print(f"  num after += 3: {num}")
num *= 2 # num = num * 2
print(f"  num after *= 2: {num}")

# In **Python**, the `f` in `print(f"...")` stands for formatted string literal, commonly called an f-string.
````

### 2.5. Control Flow (Conditionals & Loops)
Control flow statements determine the order in which instructions are executed, allowing a program to make decisions and repeat actions.
* **Conditional Statements:** `if`, `elif`, and `else` execute different blocks of code based on whether a condition is true or false.
* **Loops:** `for` and `while` repeat a block of code multiple times.

**Python:** Conditional Statements (`if`, `elif`, `else`). Execute different blocks of code based on whether a condition is true or false.


````python
score = 85

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
else:
    print("Grade: C or lower")

# **Loops (`for` and `while`):** Repeat a block of code multiple times.
#
# `for` loop (iterating over a sequence):
print("\nFor loop example:")
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(f"I like {fruit}")

# `while` loop (repeats as long as a condition is true):
print("\nWhile loop example:")
count = 0
while count < 3:
    print(f"Count is: {count}")
    count += 1 # Increment count by 1

````

### 2.6. File Input/Output (I/O)
Programs often need to interact with files to read data or save results, enabling data to persist even after the program closes.


Programs often need to interact with files to read data or save results.

* **Concept:** Persistence of data  data survives even after the program closes.
* **Benefit:** Enables working with large datasets, configuration files, and saving results.


**Python:** Conditional Statements (`if`, `elif`, `else`). Execute different blocks of code based on whether a condition is true or false.


````python
# writing to a file
file_content = "Hello, this is a test line.\nAnother line of text."
with open("my_output.txt", "w") as f: # 'w' for write mode (creates or overwrites)
    f.write(file_content)
print("File 'my_output.txt' written successfully.")

# reading from a file
with open("my_output.txt", "r") as f: # 'r' for read mode
    read_content = f.read()
print("\nContent of 'my_output.txt':")
print(read_content)
````

### 2.7. Functions
A **function** is a reusable block of code that performs a specific task. Defining functions helps organize code, make it more readable, and avoid repetition.

* In **Python**, you define a function using the `def` keyword, and the `return` statement sends a value back.


**Inputs and Outputs** in Python Functions:

* In Python, a function is a block of code that performs a specific task. Use the `def` keyword, followed by the function name, parentheses for parameters, and a colon. The code inside the function is indented. `return` sends a value back. To make it versatile, functions often have **inputs** and an **output**. Think of a function like a machine: you put something in (input), the machine does some work, and then it produces something new (output).

* **Inputs**, also known as arguments or parameters, are the values you provide to the function. They tell the function what data to work with.

    * How to provide inputs: You pass values to a function by placing them inside the parentheses () after the function's name.

    * By Position: The most common way is to match the order of the arguments in the function's definition.

    * By Name: You can also specify the argument names directly. This is often more readable, especially for functions with many arguments. For example, `ggplot(df, aes(x='x', y='y'))`.


* **Output** is the result the function returns after it has finished its work.


````python
def greet(name):
    """
    This function takes a name as input and returns a greeting message.
    This is a docstring, explaining what the function does.
    """
    return f"Hello, {name}! Welcome to coding."

# **Calling a function:**
message1 = greet("Alice")
print(message1)

message2 = greet("Bob")
print(message2)

# Function with multiple parameters and a calculation
def calculate_area_rectangle(length, width):
    """Calculates the area of a rectangle."""
    area = length * width
    return area

rect_area = calculate_area_rectangle(10, 5)
print(f"Area of rectangle (10x5): {rect_area}")
````

### 2.8 Data Visualization

Visualizing data is crucial for understanding patterns, trends, and outliers.

* **Concept:** Presenting data graphically to gain insights.
* **Benefit:** Makes complex data understandable, aids in decision-making, storytelling.

**Python:** Using `matplotlib` (a widely used plotting library) and `seaborn` (built on matplotlib, provides a high-level interface).

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data
x_values = [1, 2, 3, 4, 5]
y_values = [2, 4, 5, 4, 6]

# Line plot
plt.figure(figsize=(6, 4)) # Set figure size
plt.plot(x_values, y_values, marker='o') # 'o' for circle markers
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show() # Display the plot

# Scatter plot using Seaborn with a Pandas DataFrame
np.random.seed(42) # for reproducibility
random_data = np.random.rand(50, 2)
df = pd.DataFrame(random_data, columns=['Feature 1', 'Feature 2']) # Create a Pandas DataFrame

plt.figure(figsize=(6, 4))
sns.scatterplot(x='Feature 1', y='Feature 2', data=df)
plt.title("Seaborn Scatter Plot with Pandas DataFrame")
plt.xlabel("Feature 1 Value")
plt.ylabel("Feature 2 Value")
plt.show()
````
